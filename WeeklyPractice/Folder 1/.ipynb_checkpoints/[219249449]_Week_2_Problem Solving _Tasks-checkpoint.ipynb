```python
import pandas as pd
data= pd.read_csv('data/iris.txt', delimiter=',') 
print(data)
data.columns = ["SLength", "SWidth","PLength","PWidth","Class"]
```

```python
data.tail()
```

```python
import seaborn as sns
print(data.head())
sns.jointplot(x="SLength", y="SWidth", data=data, size=7)
```

```python
sns.jointplot(x="PLength", y="PWidth", data=data, size=7)
```

# Boxplots

```python

sns.boxplot(x="Class", y="PLength", palette="husl", data=data)
```

```python
sns.boxplot(x="Class", y="PWidth", palette="husl", data=data)
```

```python
sns.boxplot(x="Class", y="SLength", palette="husl", data=data)
```

```python
sns.boxplot(x="Class", y="SWidth", palette="husl", data=data)
```

# Line graphs

```python
import matplotlib.pyplot as plt
data['Id'] = range(1, len(data) + 1)
plt.plot(data.Id, data["SLength"], "g--")
plt.plot(data.Id, data["SWidth"], "r--")
plt.suptitle("Sepal L(Green) vs W(Red)")
plt.ylabel("cms")
plt.show()
```

```python
plt.plot(data.Id, data["PLength"], "g--")
plt.plot(data.Id, data["PWidth"], "r--")
plt.suptitle("Petal L(green) vs W(red)")
plt.ylabel("cms")
plt.show()
```

```python
plt.plot(data.Id, data["SLength"], "g--")
plt.plot(data.Id, data["PWidth"], "r--")
plt.suptitle("Sepal L(green) vs Petal W(red)")
plt.ylabel("cms")
plt.show()
```

```python
plt.plot(data.Id, data["SLength"], "g--")
plt.plot(data.Id, data["PLength"], "r--")
plt.suptitle("Sepal L(green) vs Petal L(green)")
plt.ylabel("cms")
plt.show()
```

```python
plt.plot(data.Id, data["SWidth"], "g--")
plt.plot(data.Id, data["PLength"], "r--")
plt.suptitle("Sepal W(green) vs Petal Length(red)")
plt.ylabel("cms")
plt.show()
```

```python
plt.plot(data.Id, data["SWidth"], "g--")
plt.plot(data.Id, data["PWidth"], "r--")
plt.suptitle("Sepal W(green) vs Petal W(red)")
plt.ylabel("cms")
plt.show()
```

# Brief about the findings:
From Boxplots:
sns.boxplot(x="Class", y="PLength", palette="husl", data=data): In this block of code, the data records are being plotted based on the different classes and the petal length given in form of a boxplot, the mean values, maximum value, minimum value of the petal length can be found for each of the class and compared, as iris-virginica has highest mean petal length as compared to other two classes(versicolor and setosa).
sns.boxplot(x="Class", y="PWidth", palette="husl", data=data):In this block of code, the data records are being plotted based on the different classes and the petal width given in form of a boxplot, the mean values, maximum value, minimum value of the petal width can be found for each of the class and compared, as iris-virginica has highest mean petal length as compared to other two classes(versicolor and setosa).


From Linegraphs:
The comparative values of the width of sepal or petal is less as compared to their lengths, sepal length increases, width also increases, petal length increases, width also increases.
Increase in sepal length means increase in width and decreas
When the sepal width decreases the petal width increases and vice versa.
When the sepal width increases the petal width decreases and vice versa.
