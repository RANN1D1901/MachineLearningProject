# Using the sklearn load_boston dataset and use sklearn’s regression class to learn a model. Use a range of training/test sizes using sklearn’s cross validation package. Run a 100 trials and graph the mean square error.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
data = load_boston()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
df.head()
```

```python
corr=df.corr()
corr
```

```python
data=df
# Create correlation matrix
corr_matrix = data.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=4).astype(np.bool))

# Find index of feature columns with correlation greater than 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]
display(data.shape)
# Drop correlated features 
for i in to_drop:
    data = data.drop(i, axis=1)
data.shape
feature_cols=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
       'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO',
       'B','LSTAT']
X = data[feature_cols]
y = data.target
print(type(X))
print(X.shape)
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.linear_model import LinearRegression
my_model = LinearRegression()
#fit the model using our data
my_model.fit(Xtrain, ytrain)
ypredicts=my_model.predict(Xtest)
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(ytest,ypredicts))
print('MSE:',metrics.mean_squared_error(ytest,ypredicts))
print('RMSE:',np.sqrt(metrics.mean_squared_error(ytest,ypredicts)))
print('R^2 =',metrics.explained_variance_score(ytest,ypredicts))
```

```python
#PLOT THE GRAPH OF MSE CALCULATED FOR DIFFERENT SPLITS 
from sklearn.model_selection import cross_val_score
scores =abs(cross_val_score(my_model, X, y, cv=100,scoring='neg_mean_squared_error'))
scores=pd.DataFrame(scores)
print(scores)
from matplotlib import pyplot as plt
scores.plot.bar(figsize=(30,20))
plt.title("MSE vs Trial Number")
plt.xlabel("Trial Number")
plt.ylabel("MSE")
plt.legend( title="MSE",loc=2, fontsize='small', fancybox=True)
```

```python
print(my_model.intercept_)
print(my_model.coef_)
```

# Load the full iris dataset from last week. Using 10-fold cross validation use sklearn’s LogisticRegression class to learn a model for each fold. In a Markdown cell report briefly on what level of accuracy was achieved and how it compared to the results from the decision tree induction performed last week.

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_df=pd.read_csv("data/iris.txt")
iris_df=iris_df.rename(columns={" class": "Class"})
iris_df
```

```python
from sklearn.preprocessing import LabelEncoder 
labelencoder= LabelEncoder() #initializing an object of class LabelEncoder
iris_df['Class'] = labelencoder.fit_transform(iris_df['Class']) #fitting and transforming the desired categorical column.
```

```python
feature_cols =['sepal length', ' sepal width', ' petal length', ' petal width']
X = iris_df[feature_cols]
Y = iris_df.Class
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.3)
```

```python
#IMPLEMENT THE CLASSIFIER
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1e20)
%time model.fit(Xtrain, ytrain)
```

```python
predicts=model.predict(Xtest)
```

```python
model.score(Xtest,ytest)
```

```python
#EVALUATION OF MODEL
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, Y, cv=10, scoring='accuracy')
print("Accuracy for each fold: {}".format(scores))
print("Mean Accuracy: {}".format(np.mean(scores)))
```
